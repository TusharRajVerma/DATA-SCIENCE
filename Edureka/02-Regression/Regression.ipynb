```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df = pd.read_csv('Automobile price data _Raw_.csv')
```

```python
df.head()
```

```python
df=df.replace(to_replace ="?", value =np.NaN)
df
```

```python
df=df.fillna( method ='bfill')
```

```python
#task-2 Various visualizations to understand patterns in Automobile Datasets
df['engine-size'].hist()
```

```python
#TASK-3:-Identify the errors and fix them before modelling the Automobile prices
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
df.iloc[:,5]= label_encoder.fit_transform(df.iloc[:,5])
df.iloc[:,2]= label_encoder.fit_transform(df.iloc[:,2])
df.iloc[:,3]= label_encoder.fit_transform(df.iloc[:,3])
df.iloc[:,4]= label_encoder.fit_transform(df.iloc[:,4])
df.iloc[:,6]= label_encoder.fit_transform(df.iloc[:,6])
df.iloc[:,7]= label_encoder.fit_transform(df.iloc[:,7])
df.iloc[:,8]= label_encoder.fit_transform(df.iloc[:,8])
df.iloc[:,14]= label_encoder.fit_transform(df.iloc[:,14])
df.iloc[:,15]= label_encoder.fit_transform(df.iloc[:,15])
df.iloc[:,17]= label_encoder.fit_transform(df.iloc[:,17])
df
```

```python
from sklearn.preprocessing import Normalizer

df2 = Normalizer().fit_transform(df)
df2 = pd.DataFrame(df2)
df2
```

```python
X = df2.iloc[:,:25]
```

```python
y = df2.iloc[:,25:]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2)
X_train.head()
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
```

```python
#Traininge
rgr.fit(X_train, y_train)
```

```python
# The coefficients
print('Coefficients', regr.coef_)
print('Intercept',regr.intercept_)
```

```python
#Testing
y_pred = regr.predict(X_test)
```

```python
#TASK-4 Provide a document or script which explain about model performance on what basis we need to accept your model
#Evaluation
#accuracy is 97% and mse is 0.00015
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(y_test, y_pred)) #Coefficient of Determination
print(mean_squared_error(y_test, y_pred)) #MSE
```
